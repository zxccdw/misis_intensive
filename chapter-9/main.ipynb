```python
# 4 задача 
import csv
from random import choice, shuffle

pass_letters_lower_ = "abcdefghijklmnopqrstuvwxyz"
pass_letters_upper_ = "abcdefghijklmnopqrstuvwxyz".upper()
pass_numbers_ = "0123456789"

def fio_parse(people: str):
    people = people.split()
    first_name = people[1]
    second_name = people[0]
    third_name = people[2]
    return f'{second_name}_{first_name[0]}{third_name[0]}'
    

def generate_pass():
    ans_pass = ""
    ans_pass += choice(pass_letters_lower_) + choice(pass_letters_upper_) + choice(pass_numbers_)
    for i in range(5):
        ans_pass += choice(pass_letters_lower_ + pass_letters_upper_ + pass_numbers_)
        
    ans_pass_list = list(ans_pass)
    shuffle(ans_pass_list)
    return ''.join(ans_pass_list)

    
    
    
with open("data/students.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    data = list(reader)[1::]
    
with open("data/students_new.csv", "w") as f:
    w = csv.writer(f)
    w.writerow(['id', 'Name', 'titleProject_id', 'class', 'score', 'login', 'password'])
    for row in data:
        row.append(fio_parse(row[1]))
        row.append(generate_pass())
        w.writerow(row)
    
    
# print(fio_parse(data[0][1]))
```

```python
# 3 задача 

import csv

with open("data/students.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    data = list(reader)[1::]

# print(data)
def bin_search(row_id: int):
    l = 0
    r = 500
    while r - l > 1:
        m = (l + r) // 2
        if int(data[m][0]) > row_id:
            r = m
        else:
            l = m     

    if int(data[l][0]) == row_id:
        return data[l]
    return -1
    
        
 
request = input()
ans = None

while request != "СТОП":
    ans = bin_search(int(request))
        
    if ans == -1:
        print("Ничего не найдено")
    else:
        print(f'Проект № {ans[-3]} делал: {ans[1]} он(а) получил(а) оценку - {ans[-1]}')
    
    request = input()
    ans = None

    
```
