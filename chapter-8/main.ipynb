```python
from collections import defaultdict # пример, если забыли про csv

su_classes = defaultdict(int)
cnt_classes = defaultdict(int)
avg_classes = defaultdict(float)

with open("data/students.csv", "r", encoding="UTF-8") as f:
    head = f.readline().rstrip()
    rows = []
    for row in f.readlines():
        row = row.rstrip().split(",")
        rows.append(row)
        if row[1].startswith("Хадаров Владимир"):
            # print(row)
            print(f'Ты получил: {row[-1]}, за проект - {row[-3]}')
        if row[-1] != "None":
            cnt_classes[row[-2]] += 1
            su_classes[row[-2]] += int(row[-1])
            
for classes in cnt_classes.keys():
    avg_classes[classes] = round(su_classes[classes] / cnt_classes[classes], 3)


with open("data/students_new.csv", "w", encoding="UTF-8") as f:
    print(head, file=f)
    for row in rows:
        if row[-1] == "None":
            row[-1] = avg_classes[row[-2]]
        print(*row, sep=',', file=f)
        
        
```

```python
import csv

def insertion_sort(ar):
    n = len(ar)
    for i in range(1, n):
        key = ar[i]
            
        j = i - 1
        while j >= 0 and key[-1] > ar[j][-1]:
            ar[j + 1] = ar[j]
            j -= 1
        ar[j + 1] = key
        
    return ar


with open("data/students.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    data = list(reader)[1::]

    
    sorted_answer = data.copy()
    for ind in range(len(sorted_answer)):
        if sorted_answer[ind][-1] == "None":
            sorted_answer[ind][-1] = 0
        else:
            sorted_answer[ind][-1] = int(sorted_answer[ind][-1])
            
        if '10' not in sorted_answer[ind][-2]:
            sorted_answer[ind][-1] = 0
    
    sorted_answer = insertion_sort(sorted_answer)

print("10 класс:")
for num, row in enumerate(sorted_answer[:3:], start=1, ):
    people = row[1].split()
    print(f'{num} место: {people[1][0]}. {people[0]}')
            
            
    
    
```

```python
import csv
from random import choice

pass_letters_ = "abcdefghijklmnopqrstuvwxyz" + "abcdefghijklmnopqrstuvwxyz".upper() + "0123456789"
    

def fio_parse(people: str):
    people = people.split()
    first_name = people[1]
    second_name = people[0]
    third_name = people[2]
    return f'{second_name}_{first_name[0]}{third_name[0]}'
    

def generate_pass():
    return ''.join(choice(pass_letters_) for _ in range(8))
    
    
with open("data/students.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    data = list(reader)[1::]
    
with open("data/students_new.csv", "w") as f:
    w = csv.writer(f)
    w.writerow(['id', 'Name', 'titleProject_id', 'class', 'score', 'login', 'password'])
    for row in data:
        row.append(fio_parse(row[1]))
        row.append(generate_pass())
        w.writerow(row)
    
    
# print(fio_parse(data[0][1]))
    
```

```python
import csv

x_ = 257
mod_ = 10 ** 9 + 9

def hash_st(st: str) -> int:
    h = 0
    for i in st:
        h += (h * x_ + (ord(i) - ord('а') + 1)) % mod_
    return h

    
with open("data/students.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    data = list(reader)[1::]
    
with open("data/students_new.csv", "w") as f:
    w = csv.writer(f)
    w.writerow(['id', 'Name', 'titleProject_id', 'class', 'score', 'login', 'password'])
    for row in data:
        row[0] = hash_st(row[1])
        w.writerow(row)
    
```
